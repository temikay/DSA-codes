```python
def jaccard(v1, v2):
    intersection = 0
    union = 0
    
    for a, b in zip(v1, v2):
        if a == 1 or b == 1:
            union += 1
        if a == 1 and b == 1:
            intersection += 1
            
            
    return intersection / union if union != 0 else 0
```

```python
threshold = 0.3
graph = {}

names = list(students.keys())

for i in range(len(names)):
    for j in range(i+1, len(names)):
    
        s1 = names[i]
        s2 = names[j]
    
        sim = jaccard(students[s1], students[s2])
    
        if sim >= threshold:
        
            graph.setdefault(s1, []).append((s2, sim))
            graph.setdefault(s2, []).append((s1, sim))
```

```python
for student in graph:
    print(student)
    for neighbour, weight in graph[student]:
        print("  ->", neighbour, "(", round(weight,2), ")")
```
